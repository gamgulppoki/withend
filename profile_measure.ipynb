#### 감성 분석 및 주제 분석 모듈

```python
def read_chat(chat):
    p1, p2 = chat['personas'][0], chat['personas'][1] # 사용자 id
    t1, t2 = chat['utterances'][0]['text'], chat['utterances'][1]['text'] # 사용자 발화
    return p1, p2, t1, t2
```

```python
## 각 화자별 발화를 합쳐서 하나의 대화로 return
def text_to_dialogue(text1, text2):
    text1 = text1.split('\n')
    text2 = text2.split('\n')
    dialogue = '\n'.join(item for pair in zip(text1, text2) for item in pair)
    remaining_item = text1[len(text2)] if len(text1) > len(text2) else (text2[len(text1)] if len(text1) < len(text2) else '')
    dialogue += '\n' + remaining_item
    return dialogue
```

```python
## 불용어 처리
def preprocess_stopwords(text):
    stopwords = ['키키', '진짜', '하하', 'ㅠㅠ', '나도', '크크크', '많이', '맞아', '그래', '너무', '요즘', '근데', '엄청', '나는', '같아', '그냥', '우리', '그럼', '그거', '있어', '응응', '웅웅', '정말', '지금', '아니', 'ㅜㅜ', '그리고', '내가', '그래서', '저는', '제가', '같아요', '그냥', '저도', '정말', '많이', '너무', '저희', '같이']
    for s in stopwords:
        # series = series.apply(lambda x: x.replace(s, ''))   # => input이 series일 경우
        text = text.replace(s,'')
    return text
```

```python
def extract_topic(dialogue):
    dialogue = [preprocess_stopwords(d)for d in dialogue]
    topics, probs = topic_model.transform(dialogue)
    return topics
```

```python
def save_score(id, topics, sents):
    # 사용자의 카테고리별 score 저장
    for t, s in zip(topics, sents):
        Person[id][t][0] += s
        Person[id][t][1] += 1
    return
```

#### 사용자마다 프로필 생성 (성향벡터)

```python
# 성향벡터 저장 구조 할당
def default_per_data():
    return {i: [0,0] for i in range(-1,48)}

```

```python
# 사용자 profile json 파일 경로
user_path = '/content/drive/MyDrive/dataset/personas/'
user_file_list = [os.path.join(user_path, f) for f in os.listdir(user_path)]
```

```python
# 대화 json 파일 경로
chat_path = '/content/drive/MyDrive/dataset/chats/personal/'
```

```python
Person = defaultdict(default_per_data)

for user in tqdm(user_file_list):
    with open(user, "r", encoding = "utf-8") as f:
        data = json.load(f)
    id = data["id"]
    chats = data["personal chats"]
    texts = []
    dialogues = []

    for c in chats:
        with open(chat_path+c+'.json', "r", encoding = "utf-8") as f:
            chat = json.load(f)
        p1, p2, t1, t2 = read_chat(chat)

        t = t1 if (id == p1) else t2
        texts.append(t)
        dialogues.append(text_to_dialogue(t1, t2))


    sents = extract_sentiment(texts)

    topics = extract_topic(dialogues)
    save_score(id, topics, sents)


    for topic, score_list in Person[id].items():
        if score_list[1] != 0:
            score_list[0] /= float(score_list[1])

    data["profile"] = Person[id].copy()
    with open(user, "w", encoding = "utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent = 4)
```
